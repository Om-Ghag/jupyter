# Sorting

```python
# showing pokemon according to attack
df.sort_values(['Attack'],ascending=False)
```

# making changes in data

```python
#lets make a total columns which adds all attributes of pokemon 
df['total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
df.head(5)
```

```python
#instead of using this method we can use
df['total'] = df.iloc[:,4:10].sum(axis=1) #df.iloc[row:row,col:col]
df.head(5)
```

```python
#lets change the position of total 
cols = list(df.columns)
cols
```

```python
df = df[cols[0:4]+cols[-1]+cols[4:12]] #u can also use -2 intead of 12 if u knew it was the 2nd last column
```

## solving the error 

```python
#we got the above error because python considers the cols[-1] a string and not a list
df = df[cols[0:4]+[cols[-1]]+cols[4:12]]
df.head(5)
```

```python
power = df.sort_values(['total'],ascending=False)
power.head(5)
```

## Filtering Data


```python
grass_df = df.loc[df['Type 1']=='Grass']
grass_df.head(5)
```

```python
# to get rid of the weird new index
grass_df.reset_index(drop=True,inplace=True) # we used drop = true to drop the previousindex & inplace = true to save memeory
grass_df.head()
```

```python
#lets say u want to filter out every mega pokemon
mega_df = df.loc[df['Name'].str.contains('Mega')]
#the line below is just to change index
mega_df.reset_index(drop = True,inplace = True)
mega_df.head()
```

```python
#if you want to see everythiing except mega
nonmega_df = df.loc[~df['Name'].str.contains('Mega')]
#the line below is just to change index
nonmega_df.reset_index(drop = True,inplace = True)
nonmega_df.head()

```
